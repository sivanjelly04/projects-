```python
import random

while True:
  user_choice = input("Do you want to roll the dice? (y/n): ").lower()
  if user_choice == "y":
    die1 = (random.randint(1,6))
    print(f"{die1}")
  elif user_choice == "n":
    print("Thanks for playing! ")
    break
  else:
    print("Invalid, pls try again! ")
```
