```python
import random

while True:
  choices = ("r", "p", "s")
  choice = input("Choose rock papers or scissors (r/p/s): ").lower()
  if choice not in choices:
    print("Invalid, try again! ")
    continue

  computer = random.choice(choices)

  print(f"YOU chose {choice}")
  print(f"COMPUTER chose {computer}")
  if computer == "r" and choice == "s" or computer == "p" and choice == "r" or computer == "s" and choice == "p":
    print("you lose! ")
  elif computer == choice:
    print("tied")
  else:
    print("you win! ")

  choices2 = ("y", "n")
  choice2 = input("Continue? (y/n): ").lower()
  if choice2 not in choices2:
     print("Invalid, try again! ")
  if choice2 == "n":
    print("Thanks for playing!")
    break
  elif choice2 == "y":
    continue
```
